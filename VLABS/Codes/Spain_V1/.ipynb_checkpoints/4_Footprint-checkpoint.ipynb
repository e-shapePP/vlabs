```python
# Function to add date variables to DataFrame.

def add_date_info(df):
    
    df['Timestamp'] = pd.to_datetime(df['TIMESTAMP_END'], unit='ms') 
    df['Year'] = pd.DatetimeIndex(df['Timestamp']).year
    df['Month'] = pd.DatetimeIndex(df['Timestamp']).month
    df['Day'] = pd.DatetimeIndex(df['Timestamp']).day
    df['Hour'] = pd.DatetimeIndex(df['Timestamp']).hour
    df['Minute'] = pd.DatetimeIndex(df['Timestamp']).minute
    df['DOY'] = pd.DatetimeIndex(df['Timestamp']).dayofyear
    return df
```

```python
def add_ffp_info(df):

    df['zm'] = instrument_height
    df['d']  = displacement_height
    df['z0'] = roughness_lenght    

    eu_variables          = ['TIMESTAMP_START', 'TIMESTAMP_END','FC','FC_SSITC_TEST',
                              'TA','ET','P_RAIN','RH','VPD','USTAR','SW_IN','MO_LENGTH',
                              'WS','WS_MAX','WD','V_SIGMA',
                              'FETCH_10','FETCH_30','FETCH_50','FETCH_70','FETCH_90']

    df_ec_variables       = ['Year','Month','Day','Hour','Minute',
                             'zm','d','z0','WS','MO_LENGTH','V_SIGMA','USTAR','WD']

    ffp_variables         = ['yyyy','mm','day','HH_UTC','MM',
                             'zm','d','z0','u_mean','L','sigma_v','u_star','wind_dir']

    df = df.loc[:,df_ec_variables].set_axis(ffp_variables , axis=1)

    return df
```

```python
#Sub data set with stable conditions to calculate boundary layer

df_ffp_0_2020_st = df_ffp_0_2020[df_ffp_0_2020['u_star']>0.1].copy(deep=True)   #1822
df_ffp_0_2021_st = df_ffp_0_2021[df_ffp_0_2021['u_star']>0.1].copy(deep=True)   #3636
```

```python
df_ffp_0_2020_st['h_convective'] = 1500
df_ffp_0_2021_st['h_convective'] = 1500
```

```python
#Add boundary layer data

def add_bl(df):
        #Boundary layer
        # Coriolis parameter
        angular_velocity = 7.29e-05 # angular velocity of the Earthâ€™s rotation
        latitude = 36.9985
        radianes = (latitude* math.pi)/180

        f = 2*angular_velocity*math.sin(radianes)
        df['h_st'] = df['L']/3.8*(-1+( 1 + 2.28 * df['u_star']/(f*df['L']))**0.5)
        return df
```

```python
df_ffp_0_2020_st = add_bl(df_ffp_0_2020_st)
df_ffp_0_2021_st = add_bl(df_ffp_0_2021_st)
```

```python
def stable(L): 
    if L < 10:
        bl = 0
    else:
        bl = 1
    return bl

df_ffp_0_2020_st['st'] = df_ffp_0_2020_st['L'].apply(stable)
df_ffp_0_2021_st['st'] = df_ffp_0_2021_st['L'].apply(stable) 
```

```python
def convective(L): 
    if L < 10:
        bl = 1
    else:
        bl = 0
    return bl

df_ffp_0_2020_st['convective'] = df_ffp_0_2020_st['L'].apply(convective)
df_ffp_0_2021_st['convective'] = df_ffp_0_2021_st['L'].apply(convective) 
```

```python
df_ffp_0_2020_st.replace(np.nan, -999, inplace=True) 
df_ffp_0_2021_st.replace(np.nan, -999, inplace=True) 


df_ffp_0_2020_st['h'] =  df_ffp_0_2020_st['h_convective']*df_ffp_0_2020_st['convective']
df_ffp_0_2021_st['h'] =  df_ffp_0_2021_st['h_convective']*df_ffp_0_2021_st['convective']
```

```python
df_ffp_0_2020_st['h'] =  df_ffp_0_2020_st['h'] + df_ffp_0_2020_st['h_st']*df_ffp_0_2020_st['st']
df_ffp_0_2021_st['h'] =  df_ffp_0_2021_st['h'] + df_ffp_0_2021_st['h_st']*df_ffp_0_2021_st['st'] 
```

```python
def variablesToList(df):

    zmt      = df['zm'].to_numpy().tolist()             #0
    z0t      = df['z0'].to_numpy().tolist()             #1
    u_meant  = df['u_mean'].to_numpy().tolist()         #2
    ht       = df['h'].to_numpy().tolist()              #3
    olt      = df['L'].to_numpy().tolist()              #4
    sigmavt  = df['sigma_v'].to_numpy().tolist()        #5
    ustart   = df['u_star'].to_numpy().tolist()         #6
    wind_dirt= df['wind_dir'].to_numpy().tolist()       #7
    domaint  = [-250., 250., -250., 250.]
    #nxt=1100
    rst=[20.,40.,60.,80.]
    return zmt,z0t,u_meant,ht,olt,sigmavt,ustart,wind_dirt,domaint,rst
```

```python
ffp2020 = variablesToList(df_ffp_0_2020_st)
ffp2021 = variablesToList(df_ffp_0_2021_st)
```

```python
def calculateFootprint(list):
    l = list
    FFP = FFP_climatology (zm=l[0], z0=None, umean = l[2], h=l[3], ol=l[4], sigmav=l[5],
    ustar=l[6], wind_dir=l[7], domain=l[8], nx=None, rs=l[9],smooth_data=1, fig=1)
    return FFP
```

```python
footprint2020 = calculateFootprint(ffp2020)
```

```python
pro = Proj('+proj=utm +zone=29 +ellps=GRS80 +towgs84=0,0,0,0,0,0,0 +units=m +no_defs')
#pro = Proj("epsg:25829", preserve_units=False)
latitude  = 36.9985   #728294       
longitude = -6.4345   #4097783
lat = latitude
lon = longitude
```

```python
UTMlon, UTMlat = pro(lon, lat)
#lon, lat = myProj(df['Meters East'].values, df['Meters South'].values, inverse=True)
lon_rec, lat_rec = pro(UTMlon, UTMlat , inverse=True)
```

```python
#This function is used to retrieve data from the FFP and generate dataframes with coordinates with a specific georeference
fp = footprint2020

x20 = fp['xr'][0]
y20 = fp['yr'][0]

x40 = fp['xr'][1]
y40 = fp['yr'][1]

x60 = fp['xr'][2]
y60 = fp['yr'][2]

x80 = fp['xr'][3]
y80 = fp['yr'][3]

d_20 = {'lon': x20, 'lat': y20}
d_40 = {'lon': x40, 'lat': y40}
d_60 = {'lon': x60, 'lat': y60}
d_80 = {'lon': x80, 'lat': y80}

ffp_20 = pd.DataFrame(data=d_20)
ffp_40 = pd.DataFrame(data=d_40)
ffp_60 = pd.DataFrame(data=d_60)
ffp_80 = pd.DataFrame(data=d_80)

ffp_20['UTMlon'] = UTMlon
ffp_20['UTMlat'] = UTMlat

ffp_20['lon2'] = ffp_20['lon'] + ffp_20['UTMlon'] 
ffp_20['lat2'] = ffp_20['lat'] + ffp_20['UTMlat']

ffp_40['UTMlon'] = UTMlon
ffp_40['UTMlat'] = UTMlat

ffp_40['lon2'] = ffp_40['lon'] + ffp_40['UTMlon'] 
ffp_40['lat2'] = ffp_40['lat'] + ffp_40['UTMlat']

ffp_60['UTMlon'] = UTMlon
ffp_60['UTMlat'] = UTMlat

ffp_60['lon2'] = ffp_60['lon'] + ffp_60['UTMlon'] 
ffp_60['lat2'] = ffp_60['lat'] + ffp_60['UTMlat']

ffp_80['UTMlon'] = UTMlon
ffp_80['UTMlat'] = UTMlat

ffp_80['lon2'] = ffp_80['lon'] + ffp_80['UTMlon'] 
ffp_80['lat2'] = ffp_80['lat'] + ffp_80['UTMlat']

ffp_20['lon_20'], ffp_20['lat_20'] = pro(ffp_20['lon2'].values, ffp_20['lat2'].values, inverse=True)
ffp_40['lon_40'], ffp_40['lat_40'] = pro(ffp_40['lon2'].values, ffp_40['lat2'].values, inverse=True)
ffp_60['lon_60'], ffp_60['lat_60'] = pro(ffp_60['lon2'].values, ffp_60['lat2'].values, inverse=True)
ffp_80['lon_80'], ffp_80['lat_80'] = pro(ffp_80['lon2'].values, ffp_80['lat2'].values, inverse=True)

a = ffp_20[['lon_20','lat_20']]
b = ffp_40[['lon_40','lat_40']]
c = ffp_60[['lon_60','lat_60']]
d = ffp_80[['lon_80','lat_80']]

filepath = cwd + '\\Data\\Footprint\\'
a.to_csv(filepath + "DNP_footprint_20_2020_cg.csv")
b.to_csv(filepath + "DNP_footprint_40_2020_cg.csv")
c.to_csv(filepath + "DNP_footprint_60_2020_cg.csv")
d.to_csv(filepath + "DNP_footprint_80_2020_cg.csv")
```
