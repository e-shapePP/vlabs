# Load data

```python
import pandas as pd
import datetime as dt
import numpy as np
```

```python
#This direction has to be change to the local directory with the data files

mainpath = "C:/Users/Administrador/OneDrive/Documentos/"
filepath = mainpath + "MSc Thesis/Data/"
filename_GPP = 'NEE_2020_2021/GPP/GPP_eshape.txt'                     #Computed with EC processing
filename_VI = 'Vegetation_indices/DNP_Vegetation_indices.txt'         #Computed with Time_Series_Vegetation_indices_final
filename_EV = 'Rainfall/DNP_metheorological_2016_2021.txt'            #Computed with precipitation_data
filename_EV_modis = 'Rainfall/DNP_metheorological_modis_2016_2021.txt'

parser = lambda date: dt.datetime.strptime(date, '%Y-%m-%d')
GPP = pd.read_csv(filepath + filename_GPP,parse_dates=[0], date_parser=parser, index_col = [0]) #, dtype = np.float64)
VI = pd.read_csv(filepath  + filename_VI ,parse_dates=[0], date_parser=parser, index_col = [0]) #, dtype = np.float64)
EV = pd.read_csv(filepath  + filename_EV,parse_dates=[0], date_parser=parser, index_col =  [0]) #, dtype = np.float64)
EV_modis = pd.read_csv(filepath  + filename_EV_modis,parse_dates=[0], date_parser=parser, index_col =  [0])
```

```python
GPP = GPP.dropna()
eshape_EV = pd.merge(left= GPP, right = EV ,
                     how="left", left_index = True , right_index = True) 
eshape_EV = pd.merge(left= eshape_EV, right = EV_modis,
                     how="left", left_index = True , right_index = True) 
eshape_VI = pd.merge(left= GPP, right = VI ,
                     how="left", left_index = True , right_index = True)  
```

```python
#eshape_EV = eshape_EV.dropna(how = 'all')
eshape_EV = eshape_EV.drop(columns=['P_RAIN_x','FC_0_f','RECO_1_f','RECO_2_f','P_RAIN_y','Timestamp'])
eshape_VI = eshape_VI.drop(columns=['RH','TA_f','VPD_f','SW_IN_f','ET','P_RAIN','FC_0_f','RECO_1_f','RECO_2_f'])
```

```python
correlation_EV = eshape_EV.corr()**2
correlation_VI = eshape_VI.corr()**2
```

```python
correlation_EV_1 = correlation_EV.reset_index().sort_values(by='GPP_1_f', ascending=False)
correlation_EV_1 = correlation_EV_1[['index','GPP_1_f']].set_index('index').drop(labels='GPP_2_f', axis=0)

correlation_EV_2 = correlation_EV.reset_index().sort_values(by='GPP_2_f', ascending=False)
correlation_EV_2 = correlation_EV_2[['index','GPP_2_f']].set_index('index').drop(labels='GPP_1_f', axis=0)

correlation_VI_1 = correlation_VI.reset_index().sort_values(by='GPP_1_f', ascending=False)
correlation_VI_1 = correlation_VI_1[['index','GPP_1_f']].set_index('index').drop(labels='GPP_2_f', axis=0)

correlation_VI_2 = correlation_VI.reset_index().sort_values(by='GPP_2_f', ascending=False)
correlation_VI_2 = correlation_VI_2[['index','GPP_2_f']].set_index('index').drop(labels='GPP_1_f', axis=0)
```

```python
Top_ev = []
for i in range (1,5,1):
    a  = correlation_EV_2.drop(labels=['Gpp_modis']).reset_index()
    a  = a['index'][i]
    Top_ev.append(a)
    print(Top_ev)
```

```python
Top_vi = []
for i in range (1,5,1):
    a  = correlation_VI_2.reset_index()
    a  = a['index'][i]
    Top_vi.append(a)
    print(Top_vi)
```

```python
eshape_EV
eshape_VI = eshape_VI.drop(labels=['GPP_1_f','GPP_2_f'], axis=1)
eshape_Full = pd.merge(left= eshape_EV, right = eshape_VI,
                     how="left", left_index = True , right_index = True)  
```

```python
eshape_Full = eshape_Full.drop(labels=['TH'], axis=1)
```

```python
a = eshape_EV.drop(labels=['GPP_1_f','GPP_2_f','TH'], axis=1).columns.to_list()
b = eshape_VI.columns.to_list()

for i in a:
    for ii in b:
        str = i+ '*'+ii
        eshape_Full[str] = eshape_Full[i]*eshape_Full[ii]
        
for i in b:
    for ii in b:
        str = i+ '*'+ii
        eshape_Full[str] = eshape_Full[i]*eshape_Full[ii]    
```

```python
for i in Top_ev:
    for ii in Top_vi:
        str = i+'*'+ii
        eshape_Full[str] = eshape_Full[i]*eshape_Full[ii]
        
```

```python
from sklearn.model_selection import train_test_split
```

```python
training_eshape, testing_eshape = train_test_split(eshape_Full, test_size = 0.3, random_state=7) #0.001 for full dataset
#training, testing = train_test_split(ex1_model_2, test_size = 0.2)
```

```python
import statsmodels.formula.api as smf
```

```python
dependant = 'GPP_1_f'
independant = 'Gpp_modis'

ex1_model_training = training_eshape[[dependant,independant]].rename(columns={
    dependant: 'GPP',
    independant: 'Gpp_modis'
})

ex1_model_test = testing_eshape[[dependant,independant]].rename(columns={
    dependant: 'GPP',
    independant: 'Gpp_modis'
})

lm_0 = smf.ols(formula="GPP~Gpp_modis", data = ex1_model_training).fit()
lm_0.pvalues
lm_0.rsquared
lm_0.summary()
```

```python
model_pred = lm_0.predict(ex1_model_test)
SSD = sum((ex1_model_test['GPP']-model_pred)**2)
SSD
RSE = np.sqrt(SSD/(len(ex1_model_test)-2))
RSE
gpp_m = np.mean(ex1_model_test['GPP'])
gpp_m 
error = RSE/gpp_m 
error
```
