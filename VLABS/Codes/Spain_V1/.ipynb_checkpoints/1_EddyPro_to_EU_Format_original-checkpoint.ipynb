# Load In-situ EC data from Duque Fuente Station.

```python
import pandas as pd
import datetime as dt
import os
```

```python
cwd  = os.getcwd()
os.chdir("..")
uwd  = os.getcwd()
os.chdir(cwd)
```

```python
#Directory with the Data
filepath = cwd + '\\Data\\'

#Create a list of dates to batch the files. Requires two periods because the data 
#between '2021-02-03' and '2021-02-11' is unavailable.

def list_dates(start, end):
    
    period_range = pd.date_range(start=start, end=end, freq="D")
    period_df    = pd.DataFrame(period_range)
    
    def str_date(timedate): 
        timestring =  timedate.strftime("%Y-%m-%d")
        return timestring 
    
    period_list = period_df[0].apply(str_date).tolist() 
    
    return period_list 

    
period_1_list = list_dates('2020-10-1', '2021-02-03')
period_2_list = list_dates('2021-02-11','2021-06-08')

time_list = period_1_list + period_2_list

#Load first file. First file of the data set is required to start the batch 

first_file = "NEE\\2020-09-30_smart3-00433_EP-Summary.txt"

df_EddyPro  = pd.read_csv(filepath + first_file , sep="\t")
length_file = len(df_EddyPro)

#Iterate dates to load the full dataset

for time in time_list:
    file = filepath + "NEE/" + time + "_smart3-00433_EP-Summary.txt"
    df = pd.read_csv(file, sep="\t")
    length_file += len(df)
    
    df_EddyPro = pd.concat([df_EddyPro, df], axis = 0).drop(index=0, axis=0)
    
    print('The length of your data set up to '+ time +' is  : ' + str(length_file))
```

# Format Eddy Pro file into European Fluxes Database Cluster file

```python
df = df_EddyPro

df['date_end']   = pd.to_datetime(df['date']+' '+df['time'])
df['date_start'] = df['date_end'] - dt.timedelta(minutes = 30)

#Select relevant variables for the model production and change labels to the European Fluxes Database Cluster format
#Global radiation 'SW_IN' is not available in the data set 

EddyPro_variables     = ['date_start','date_end','co2_flux','qc_co2_flux',
                         'air_temperature','RH','VPD','u*','L',
                         'wind_speed','max_wind_speed','wind_dir','v_var']

eu_variables          = ['TIMESTAMP_START', 'TIMESTAMP_END','FC','FC_SSITC_TEST',
                        'TA','RH','VPD','USTAR','MO_LENGTH',
                        'WS','WS_MAX','WD','V_SIGMA']


df_eu = df[EddyPro_variables].set_axis(eu_variables , axis=1)


#Transform objects in the data frames from strings into float variables

numeric_variables     = ['FC','FC_SSITC_TEST',
                        'TA','RH','VPD','USTAR','MO_LENGTH',
                        'WS','WS_MAX','WD','V_SIGMA']

for var in numeric_variables:
    df_eu[var] = pd.to_numeric(df_eu[var], errors='coerce', downcast ='float')
    
#Integrate Global Radiation Data from PVGIS-SARAH 

parser   = lambda date: dt.datetime.strptime(date, '%Y-%m-%d %H:%M:%S')
df_SW_IN = pd.read_csv(filepath + "Global radiation\\Global_Radiation_2020_2021_from_mean.txt",parse_dates=[0], date_parser=parser)
#df_SW_IN = pd.read_csv(filepath + "Global radiation\\Global_Radiation_2020_2021_cte_from_2012_2013.txt",parse_dates=[0], date_parser=parser)
#df_SW_IN = pd.read_csv(filepath + "Global radiation\\Global_Radiation_2020_2021_cte_from_2012_2013.txt",parse_dates=[0], date_parser=parser)


df_eu = pd.merge(left= df_eu, right = df_SW_IN,
                        how="left", left_on = 'TIMESTAMP_END', right_on ='TIMESTAMP_END')

#Calculate 'V_SIGMA'
df_eu['V_SIGMA'] = df_eu['V_SIGMA'].pow(0.5)


#Create a data frame with constant frequency of 30 minutes in the period of study to identify gaps in time series  


#df_time  = pd.date_range(start='2020-10-01 00:00:00', end='2021-06-8 00:00:00', freq="30min")
df_time  = pd.date_range(start='2020-09-30 12:00:00', end='2021-06-08 06:00:00', freq="30min")
df_time  = pd.DataFrame(df_time)

df_time ['TIMESTAMP_START'] = df_time [0]
df_time ['TIMESTAMP_END']   = df_time [0] + dt.timedelta(minutes = 30)
df_time.drop(columns=0,  inplace=True)

df_eu.rename(columns={'TIMESTAMP_END': 'MISSING_DATA'},inplace=True)

df_analysis = pd.merge(left= df_time, right = df_eu,
                        how="left", left_on = 'TIMESTAMP_START' , right_on ="TIMESTAMP_START")

#Create 'FC' columns with different quality check threshols

df_analysis_flag_0 = df_analysis[(df_analysis['FC_SSITC_TEST'] == 0)] #5937    High quality data
df_analysis_flag_1 = df_analysis[(df_analysis['FC_SSITC_TEST'] <= 1)] #9347    Medium quality data set
df_analysis_flag_2 = df_analysis[(df_analysis['FC_SSITC_TEST'] <= 2)] #11313   Data set with poor data

df_analysis_flag_0 = df_analysis_flag_0[['TIMESTAMP_START','FC']].rename(columns={'FC': 'FC_0'})
df_analysis_flag_1 = df_analysis_flag_1[['TIMESTAMP_START','FC']].rename(columns={'FC': 'FC_1'})
df_analysis_flag_2 = df_analysis_flag_2[['TIMESTAMP_START','FC']].rename(columns={'FC': 'FC_2'})


df_eshape = pd.merge(left= df_analysis, right = df_analysis_flag_0,
                        how="left", left_on = 'TIMESTAMP_START' , right_on ="TIMESTAMP_START")

df_eshape = pd.merge(left= df_eshape, right = df_analysis_flag_1,
                        how="left", left_on = 'TIMESTAMP_START' , right_on ="TIMESTAMP_START")

df_eshape = pd.merge(left= df_eshape, right = df_analysis_flag_2,
                        how="left", left_on = 'TIMESTAMP_START' , right_on ="TIMESTAMP_START")

#Fill null values with -9999

df_eshape = df_eshape.dropna(subset=['TIMESTAMP_START', 'TIMESTAMP_END'])
df_eshape = df_eshape.drop('MISSING_DATA', axis =1)
df_eshape = df_eshape.fillna(-9999.) 

#Export file 
df_eshape = df_eshape.set_index ('TIMESTAMP_START')
df_eshape.to_csv(filepath + "NEE\\NEE_EU_format_2020_2021.txt")

date = dt.datetime(2021, 1, 1)
df_eshape_2020 = df_eshape[(df_eshape['TIMESTAMP_END'] < date)]
df_eshape_2021 = df_eshape[(df_eshape['TIMESTAMP_END'] >= date)]
df_eshape_2020.to_csv(filepath + "NEE\\NEE_EU_format_2020.txt")
df_eshape_2021.to_csv(filepath + "NEE\\NEE_EU_format_2021.txt")
```

# Plot Net Ecosystem Productivity in gC+1m-2day-1

```python
import matplotlib.pyplot as plt
```

```python
df_NEP = df_analysis.loc[:,['TIMESTAMP_END','FC']]

df_NEP['TIMESTAMP_END']  = df_NEP['TIMESTAMP_END'].dt.date
df_NEP['NEP_grams']      = df_NEP['FC']*-12/1000000*1800

df_NEP = df_NEP.groupby('TIMESTAMP_END').sum() 
df_NEP.head()
```

```python
#To include figures in the notebook
%matplotlib inline
fig, ax = plt.subplots(figsize=(14, 6))
ax.plot(df_NEP['NEP_grams'], marker='.', ms=6, color='blue', label='2020-10-1 to 2020-06-8')
ax.set_ylim(-10, 10)
```
