# Load global radiation data

```python
import pandas as pd
import datetime as dt
import os
```

```python
cwd  = os.getcwd()
os.chdir("..")
uwd  = os.getcwd()
os.chdir(cwd)

#Directory with the Data
filepath = cwd + '\\Data\\Global radiation\\'
filename = 'Timeseries_36.984_-6.432_SA_2a_2005_2016.csv'

parser = lambda date: dt.datetime.strptime(date, '%Y%m%d:%H%M')
df_SW_IN = pd.read_csv(filepath + filename,parse_dates=[0], date_parser=parser)
```

# Format PVGIS-SARAH files into European Fluxes Database Cluster file

* Mean value

```python
df = df_SW_IN.copy()

#Delay time 9 minutes

df['time'] = df['time'] - dt.timedelta(minutes = 9)
```

```python
# Function to add date variables to DataFrame

def add_date_info(df):
    
    df['Timestamp'] = pd.to_datetime(df['time']) 
    df['leap']      = pd.DatetimeIndex(df['Timestamp']).is_leap_year
    df['Year']      = pd.DatetimeIndex(df['Timestamp']).year
    df['Month']     = pd.DatetimeIndex(df['Timestamp']).month
    df['Day']       = pd.DatetimeIndex(df['Timestamp']).day
    df['Hour']      = pd.DatetimeIndex(df['Timestamp']).hour
    df['minute']    = pd.DatetimeIndex(df['Timestamp']).minute
    df['DOY']       = pd.DatetimeIndex(df['Timestamp']).dayofyear
    
    return df

df = add_date_info(df)

#Delay timeseries 8 years to match 2012 with 2020. The assumtion is that the global radiation in a specific hour in a day of the year to differenciate day and night is a cyclical phenomena

df[['Year','Month','Day','Hour','minute']] = df[['Year','Month','Day','Hour','minute']].astype('string')
df['MONTH_DAY'] = df['Month']+'-'+df['Day']+' '+df['Hour']+':'+df['minute']

df_mean_leap     = df[df['leap']==True].groupby('MONTH_DAY').mean().reset_index()
df_mean_noleap   = df[df['leap']==False].groupby('MONTH_DAY').mean().reset_index()

df_mean_leap['date']   = '2020'+'-'+ df_mean_leap['MONTH_DAY']
df_mean_noleap['date'] = '2021'+'-'+ df_mean_noleap['MONTH_DAY']

df_mean_leap['TIMESTAMP_END'] = pd.to_datetime(df_mean_leap['date'], infer_datetime_format=True)  
df_mean_noleap['TIMESTAMP_END'] = pd.to_datetime(df_mean_noleap['date'], infer_datetime_format=True)  

df_mean_leap   = df_mean_leap.sort_values(by=['TIMESTAMP_END'])
df_mean_noleap = df_mean_noleap.sort_values(by=['TIMESTAMP_END'])


df_SW_2020_2021 = pd.concat([df_mean_leap, df_mean_noleap])

df = df_SW_2020_2021

#Rename global radiation to match the EU format

df.rename(columns={'G(i)': 'SW_IN'}, inplace=True)

#Create time series with 30 min frequency

time_series = pd.date_range(start='2020-01-01 00:00:00', end='now', freq="30min")
time_series = pd.DataFrame(time_series).rename(columns={0: 'TIMESTAMP_END'})

df_30min = pd.merge(left= time_series, right = df[['SW_IN','TIMESTAMP_END']],
                     how="left", left_on = 'TIMESTAMP_END' , right_on ='TIMESTAMP_END')

df_30min = df_30min.interpolate(method='linear', limit_direction ='backward')

#Export file

global_radiation  = df_30min.set_index ('TIMESTAMP_END')
global_radiation.to_csv(filepath + "Global_Radiation_2020_2021_from_mean.txt")
```

```python
dff = global_radiation.copy()
#dff.loc[dff['SW_IN'] < 20, 'SW_IN'] = 0
dff.loc[dff['SW_IN'] >= 20, 'SW_IN'] = 20


dff.to_csv(filepath + "Global_Radiation_2020_2021_cte_from_mean_20.txt")
```

```python
dff = global_radiation.copy()
dff.loc[dff['SW_IN'] < 20, 'SW_IN'] = 0
dff.loc[dff['SW_IN'] >= 20, 'SW_IN'] = 1


dff.to_csv(filepath + "Global_Radiation_2020_2021_cte_from_mean_1.txt")
```

* 2012-2013 values

```python
df = df_SW_IN.copy()

#Delay time 9 minutes

df['time'] = df['time'] - dt.timedelta(minutes = 9)
```

```python
# Function to add date variables to DataFrame

def add_date_info(df):
    
    df['Timestamp'] = pd.to_datetime(df['time']) 
    df['Year']      = pd.DatetimeIndex(df['Timestamp']).year
    df['Month']     = pd.DatetimeIndex(df['Timestamp']).month
    df['Day']       = pd.DatetimeIndex(df['Timestamp']).day
    df['Hour']      = pd.DatetimeIndex(df['Timestamp']).hour
    df['minute']    = pd.DatetimeIndex(df['Timestamp']).minute
    return df

df = add_date_info(df)

#Delay timeseries 8 years to match 2012 with 2020. The assumtion is that the global radiation in a specific hour in a day of the year to differenciate day and night is a cyclical phenomena

df['Year'] = df['Year'] + 8

df[['Year','Month','Day','Hour','minute']] = df[['Year','Month','Day','Hour','minute']].astype('string')
df['TIMESTAMP_END'] = df['Year']+'-'+df['Month']+'-'+df['Day']+' '+df['Hour']+':'+df['minute']

def stringToDate(time):
    datetype = dt.datetime.strptime(time, '%Y-%m-%d %H:%M')
    return datetype

df['TIMESTAMP_END'] = df['TIMESTAMP_END'].apply(stringToDate) 

#Rename global radiation to match the EU format

df.rename(columns={'G(i)': 'SW_IN'}, inplace=True)

#Create time series with 30 min frequency

time_series = pd.date_range(start='2020-01-01 00:00:00', end='now', freq="30min")
time_series = pd.DataFrame(time_series).rename(columns={0: 'TIMESTAMP_END'})

df_30min = pd.merge(left= time_series, right = df[['SW_IN','TIMESTAMP_END']],
                     how="left", left_on = 'TIMESTAMP_END' , right_on ='TIMESTAMP_END')

df_30min = df_30min.interpolate(method='linear', limit_direction ='backward')

#Export file

global_radiation  = df_30min.set_index ('TIMESTAMP_END')
global_radiation.to_csv(filepath + "Global_Radiation_2020_2021_from_2012_2013.txt")
```

* Constant value

```python
dff = global_radiation.copy()
dff.loc[dff['SW_IN'] > 20, 'SW_IN'] = 20
#dff.loc[dff['SW_IN'] < 20, 'SW_IN'] = 0

dff.to_csv(filepath + "Global_Radiation_2020_2021_cte_from_2012_2013.txt")
```

```python
data_rad = df
data_rad['Date'] = df['TIMESTAMP_END'].dt.date
data_rad_day = data_rad[['Date','SW_IN']].groupby('Date').mean().reset_index()

# Function to add date variables to DataFrame.
def add_date_info(df):
    
    df['Timestamp'] = pd.to_datetime(df['Date']) 
    df['Year'] = pd.DatetimeIndex(df['Timestamp']).year
    df['Month'] = pd.DatetimeIndex(df['Timestamp']).month
    df['Day'] = pd.DatetimeIndex(df['Timestamp']).day
    df['DOY'] = pd.DatetimeIndex(df['Timestamp']).dayofyear
    return df

data_rad_day = add_date_info(data_rad_day)
data_rad_day = data_rad_day.drop(columns=['Date'])
```
